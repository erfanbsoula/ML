```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 1000
np.random.seed(1)

X = np.concatenate([
    np.random.normal(0, 1, int(0.3 * N)),
    np.random.normal(5, 1, int(0.7 * N))
])

X = X[:, np.newaxis]
```

```python
X.shape
```

```python
class Parzen:
    def __init__(self, Data, kernel):
        self.X = Data
        self.kernel = kernel
        self.d = X.shape[1]
        self.N = X.shape[0]

    def set_new_kernel(self, kernel):
        self.kernel = kernel

    def point_estimate(self, x, kernel_size):
        return np.sum(self.kernel((self.X - x)/kernel_size)) / self.N / (kernel_size**self.d)

    def pdf(self, xs, kernel_size):
        p = [self.point_estimate(xs[i], kernel_size) for i in range(xs.shape[0])]
        return np.array(p)
```

```python
def RectKernel(X):
    return ((-0.5 < X) * (X < 0.5)).astype(np.float64)
```

```python
estimator = Parzen(X, RectKernel)
```

```python
xs = np.arange(-4, 10, 0.2)[:, np.newaxis]
ps = estimator.pdf(xs, 0.5)

plt.plot(xs.squeeze(), ps)
plt.ylim(bottom=0)
plt.show()
```

```python
def GaussianKernel(X):
    return np.exp(-X**2 / 2) / (2*np.pi)**(X.shape[1]/2)
```

```python
estimator.set_new_kernel(GaussianKernel)
xs = np.arange(-4, 10, 0.1)[:, np.newaxis]
ps = estimator.pdf(xs, 0.1)

plt.plot(xs.squeeze(), ps)
plt.ylim(bottom=0)
plt.show()
```
